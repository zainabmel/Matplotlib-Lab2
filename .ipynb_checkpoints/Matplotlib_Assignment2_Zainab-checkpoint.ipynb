# Q2: Create a one line chart using two arrays. The first array represents x and the second array represents y
| x     |     y    |
| ----- | -------- |
| 1    | 1       |
| 2    | 2       |
| 3    | 3       |
| 4    | 5       |
| 5    | 8       |
| 6    | 13      |
| 7    | 20      |

```python
# write your code here ^_^
x = [1,2,3,4,5,6,7]
y = [1,2,3,5,8,13,20]
  
plt.figure(figsize=(5,3))

plt.plot(x,y, c ='cornflowerblue', linewidth = 2)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data of x and y')
plt.show()
```

# Q3: Create a two lines chart using the same previous arrays. 
But this time:
- Mark each point with a circle.
- Use dashed lines.
- Set the size of the marker (circle) to size 10.
- Set the width of the line to 3.
- Add legends. 'Normal' for array x and 'Fast' for array y.

```python
# write your code here ^_^

plt.figure(figsize=(5,3))


plt.plot(x, c ='cornflowerblue', linewidth= 3, marker = 'o', linestyle = 'dashed', markersize = 8, markeredgewidth= 2, markeredgecolor= 'red');

plt.plot(y, c ='orange', linewidth= 3, marker = 'o', linestyle = 'dashed', markersize = 8, markeredgewidth= 2, markeredgecolor= 'red');

plt.title('Data of x and y')
plt.legend(['Normal','Fast'], shadow = True, loc = 0)
plt.show()
```

# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title 
- Add a super title 'my data visulaization assignment'. 

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])


#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])
```

```python
# write your code here ^_^
plt.figure(figsize=[25,10])
plt.suptitle('my data visulaization assignment',fontsize = 20)

#first
plt.subplot(1,3,1) 

plt.plot(x_plot1, y_plot1, c ='cornflowerblue', linewidth = 2)
plt.xlabel('plot 1 x')
plt.ylabel('plot 1 y')
plt.title('plot 1 line chart')

#second
plt.subplot(1,3,2)

plt.plot(x_plot2, y_plot2, c ='orange', linestyle = 'dashed', linewidth = 2)
plt.xlabel('plot 2 x')
plt.ylabel('plot 2 y')
plt.title('plot 2 line chart')

#third
plt.subplot(1,3,3)

plt.plot(x_plot3, y_plot3, c ='blue', linestyle = 'dotted', linewidth = 2)
plt.xlabel('plot 3 x')
plt.ylabel('plot 3 y')
plt.title('plot 3 line chart')
```

# Q5: Define an array that has a values of your top three TV series and define another array to hold the rating of the three TV series out of 10. Create a bar chart using the two arrays
- Set a new color for the bars.
- Change the width of the bars to width 0.5.

```python
# write your code here ^_^
topSeries = ["the queen's gambit","jurassic world","hunger games"]
ratings=[8.6, 6.9, 7.2]

plt.bar(topSeries, ratings, color = ["cornflowerblue", "orange", "cyan"], width=0.5)
plt.xlabel('TV Series')
plt.ylabel('Ratings')
plt.title('Top TV Series Ratings')
plt.show()
```

# Q6: Create a scatter plot  using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot
| carAge | carSpeed |
| ----- | -------- |
| 5    | 99       |
| 7    | 86       |
| 8    |87      |
| 7    | 88       |
| 2    | 111       |
| 17    | 86       |
| 2    | 103      |
| 9    | 87       |
| 4    | 94       |
| 11    | 78      |

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
# write your code here ^_^

carAge = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
carSpeed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78]

plt.scatter(carAge, carSpeed, s = 350, color = ['cornflowerblue','hotpink','black','gray','green', 'red', 'purple','orange','brown','cyan'], alpha=0.4) # s >> size

plt.legend(loc = 0)
plt.title('Car Age and Speed',fontsize = 20)
plt.xlabel('Car Age',fontsize = 20)
plt.ylabel('Car Speed',fontsize = 20)

plt.show()
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library
| python_libraries |   
| ------------ | 
| NumPy        | 
| Pandas       | 
| Matplotlib   | 
| Seaborn      | 
| Plotly       | 
| Scikit-learn | 

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
# write your code here ^_^

python_libraries = ['NumPy','Pandas','Matplotlib','Seaborn','Plotly','Scikit-learn']
skills = [9,9,6.5,4,2,7.9]

plt.figure(figsize=(5,3))

plt.title('Number of Orders every Year')
plt.pie(skills,labels=python_libraries, autopct='%1.1f%%', startangle=90, shadow=True, explode = [0.3,0,0,0,0,0], colors=['cornflowerblue','orange','hotpink','cyan','pink','blue'])

plt.show()
```

# Q8: Create a histogram with 250 values and 15 bars

```python
# write your code here ^_^

x = np.random.normal(170, 10, 250) # 250 values around 170 and 10 standard deviation

plt.hist(x, facecolor = 'cornflowerblue', edgecolor = 'hotpink', bins = 15) # bins = 15 means 15 bars
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Distribution of Random Values')
plt.show()
```
